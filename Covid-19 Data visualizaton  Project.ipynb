```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mp
import seaborn as sns
```

# Reading Data

```python
read_covid = pd.read_csv('covid.csv')
```

```python
print(read_covid)
```

# Calculating total Cases

```python
total_cases = read_covid['New_cases'].sum()
print("The Total cases aroung the globe were : ",total_cases)
```

```python
covid_frame = pd.DataFrame(read_covid)
print(covid_frame)
```

# Reading Dates of Data

```python
covid_date = pd.to_datetime (read_covid['Date_reported'])
print(covid_date)
```

```python
monthly_data = read_covid.groupby(covid_date.dt.year)[['New_cases','New_deaths']].sum()
print("The new cases and new deaths for each month: ")
print(monthly_data)
```

# Asian 3 Countries

## Afghanistan

```python
Afg = read_covid[1:223]
print(Afg)
```

```python
monthly_afg = Afg.groupby(covid_date.dt.year)[['New_cases','New_deaths']].sum()
print('\n\n',monthly_afg)
```

```python
plt.plot(monthly_afg['New_cases'])
plt.plot(monthly_afg['New_deaths'])
plt.ylabel('Number of Deaths')
plt.xlabel("Months")
plt.title("Graph shows Number of Deaths over Years in Afghanistan")
red_patch = mp.Patch(color = "orange", label = 'Total Deaths')
blue_patch = mp.Patch(color = "blue", label = 'Total Cases')
plt.legend(handles = [red_patch,blue_patch])
plt.show()
```
